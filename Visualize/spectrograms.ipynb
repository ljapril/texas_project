```python
import os
import glob
import matplotlib.pyplot as plt
from obspy import read
from obspy import UTCDateTime

def plot_spectrogram(sac_file, start_time_seconds=0, duration_seconds=3600, downsample_factor=10):
    # Read the SAC file using ObsPy
    st = read(sac_file)

    # Get the first Trace from the Stream
    tr = st[0]

    # Set the time window
    start_time = tr.stats.starttime + start_time_seconds
    end_time = start_time + duration_seconds

    # Trim the Trace data to the specified time window
    tr.trim(start_time, end_time)

    # Downsample the Trace data
    tr.decimate(factor=downsample_factor)

    # Plot the spectrogram using Matplotlib
    fig, ax = plt.subplots()
    tr.spectrogram(log=True, axes=ax)
    ax.set_title(f"Spectrogram of {sac_file}")
    ax.set_xlabel("Time [s]")
    ax.set_ylabel("Frequency [Hz]")
    plt.show()
```

```python
sac_file = "/Users/marcgarcia/Research/ElPaso_data/101/20210605.101.HHE.SAC"
start_time_seconds = 3600  # Change this value to the desired start time in seconds
duration_seconds = 7200  # Change this value to the desired duration in seconds (3600 seconds = 1 hour)

plot_spectrogram(sac_file, start_time_seconds, duration_seconds, downsample_factor=10)

```
